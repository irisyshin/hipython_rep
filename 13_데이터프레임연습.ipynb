```python
#country, continent, year 집계가능 속성
#mean, max, min

# 집계 가능 속성들
#df[['col1', 'col2']].agg(['함수1', '함수2', ...])

df1[['lifeExp','pop','gdpPercap']].agg(['mean','max','min'])
gb_contry = df1.groupby('country')[['lifeExp','pop','gdpPercap']].agg(['mean','max','min'])
gb_continent = df1.groupby('continent')[['lifeExp','pop','gdpPercap']].agg(['mean','max','min'])
gb_year = df1.groupby('year')[['lifeExp','pop','gdpPercap']].agg(['mean','max','min'])
gb_contry
```
