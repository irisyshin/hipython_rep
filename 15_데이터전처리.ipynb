```python
weather_df = pd.read_csv('data/weather.csv')
weather_df.head(), weather_df.columns
```

```python
# day > unpivot melt()

weather_melt = weather_df.melt(
    id_vars=['id','year','month','element'],
    var_name ='day',
    value_name='temp'
)

print(weather_melt)
```

```python
#2) element pivoting -> tmax,tmin
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
)

weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat
```

```python
# chaining
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
).reset_index()

weather_tidy.head()
```

# 월급 계산


```python
import pandas as pd

# 직원 명단 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 45, 35, 32, 28],
    'Salary': [5000, 7000, 6500, 5500, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python
def age_category(age):
    if age <30:
        return '20대'
    elif age<40:
        return '30대'
    else:
        return '40대 이상'
        

#category column 추가 
category = df['Age'].apply(age_category)
df['Age_category'] = category

df
```

```python
def salary_boundary(salary):
    if salary<6000:
        return '돈 아껴서!!!'
    elif salary<7000:
        return '열심히 일헀군'
    else:
        return '사고싶은 거 다 사'
```

```python
# 보너스

def calc_bonus(salary):
    return salary*0.50

df['Bonus'] = df['Salary'].apply(calc_bonus)
print(df)

# lambda
df['Bonus'] = df['Salary'].apply((lambda salary: salary*0.5))
print(df)
```

# 병합

```python
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
```

```python
pd.concat([df1,df2,df3])
```

```python
# inner> 같은 애를 데리고 와
pd.concat([df1,df2,df3], join='inner')
```

```python
# 열방향으로 병합 > index
df1.index = [0, 1, 2, 3]
df2.index = [4, 5, 6, 7]
df3.index = [0, 2, 5, 7]

col_concat = pd.concat([df1, df2, df3], axis=1)
col_concat
```

# 열병합 연습 


```python

import pandas as pd

# 월별 매출 데이터
january = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02'],
    'Sales': [200, 250]
})

february = pd.DataFrame({
    'Date': ['2023-02-01', '2023-02-02'],
    'Sales': [300, 350]
})
```

```python
monthly_sales = pd.concat([january, february], ignore_index=True)
print(monthly_sales)
print('-'*50)
january['Month'] = 'January'
february['Month'] = 'February'

monthly_sales = pd.concat([january, february], ignore_index=True)
print(monthly_sales)
```

```python
# 상점 A의 매출 데이터
store_a = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02'],
    'Sales_A': [200, 250]
})

# 상점 B의 매출 데이터
store_b = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02'],
    'Sales_B': [300, 350]
})
```

```python
concat_df = pd.concat([store_a, store_b.drop(columns='Date')], axis=1)
concat_df
```

```python
#merge
merged_df = pd.merge(store_a, store_b, on='Date')
merged_df
```

```python
#상점 C의 매출 데이터
store_c = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-03'],
    'Sales_C': [220, 280]
})

#상점 D의 매출 데이터
store_d = pd.DataFrame({
    'Date': ['2023-01-02', '2023-01-03'],
    'Sales_D': [330, 370]
})
```

```python
pd.concat([store_c.set_index('Date'),store_d.set_index('Date')],axis=1, join='outer')
```
