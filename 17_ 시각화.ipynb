```python

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6))
plt.title('Stock Priced Over 5 Days')
plt.ylim(95,110)
plt.grid()
plt.xlabel('Date')
plt.ylabel('Price')

plt.plot(dates, prices, label='Stock Price')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1) # 3rd parameter 번째 그림
plt.plot(dates, prices, color='Green') #  line chart
plt.xticks(rotation=40)

plt.subplot(1,2,2)
plt.bar(dates, prices, color='Orange')
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart')
```
