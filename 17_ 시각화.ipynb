```python

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6))
plt.title('Stock Priced Over 5 Days')
plt.ylim(95,110)
plt.grid()
plt.xlabel('Date')
plt.ylabel('Price')

plt.plot(dates, prices, label='Stock Price')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1) # 3rd parameter 번째 그림
plt.plot(dates, prices, color='Green') #  line chart
plt.xticks(rotation=40)

plt.subplot(1,2,2)
plt.bar(dates, prices, color='Orange')
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart')
```

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df
```

```python
# 1별량 그래프 : 히스토그램
plt.hist(data=tips_df, x='total_bill')
```

```python
# 1별량 그래프 : 히스토그램
fig = plt.figure(figsize=(12,5))
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)

axes1.hist(data=tips_df, x='total_bill', bins = 20)
axes2.hist(data=tips_df, x='tip', bins=20)
plt.tight_layout()
plt.show()

# x = total bill, tip
# y = 빈도

# 팁은  $2-2.5 을 주는 사람이 가장 빈도수가 높다 
# $10-20 사이가 빈도수가 가장 높다 
# 이상치가 얼마 있는지 파악가능 
```

```python
# box plot 
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(
    x=[
        tips_df[tips_df['sex']=='Female']['total_bill'],
        tips_df[tips_df['sex']=='Male']['total_bill']
    ],
    label=['Female','Male']
)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(
    x=[
        tips_df[tips_df['sex']=='Female']['tip'],
        tips_df[tips_df['sex']=='Male']['tip']
    ],
    labels=['Female','Male']
)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(
    x=[
        tips_df[tips_df['sex']=='Female']['size'],
        tips_df[tips_df['sex']=='Male']['size']
    ],
    labels=['Female','Male']
)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(
    x=[
        tips_df[tips_df['size']==2]['tip'],
        tips_df[tips_df['size']==4]['tip']
    ],
    labels=['2','4']
)
plt.show()
```

```python
# 토,일요일에 팁 금액이 가장 많고 분산도 큼 → 손님 많거나 팁을 더 후하게 주는 경향있음 
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)
axes.boxplot(
    x=[
        tips_df[tips_df['day']=='Thur']['total_bill'],
        tips_df[tips_df['day']=='Fri']['total_bill'],
        tips_df[tips_df['day']=='Sat']['total_bill'],
        tips_df[tips_df['day']=='Sun']['total_bill']
    ],
    labels=['Thur','Fri','Sat','Sun']
)
plt.title(' 요일에 따른 Total Bill의 변화')
plt.show()
```
