# 3. ë°ì´í„° ì „ì²˜ë¦¬

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

* ë‹¨ìˆœíˆ 1ë¶€í„°ì˜ ìˆ«ìë¥¼ ë¶€ì—¬í•œ 'cust_id'ë¥¼ ìˆ˜ì¹˜í˜• ë³€ìˆ˜ë¡œ ë°›ì•„ë“¤ì´ë©´, ê²°ê³¼ê°€ ì™œê³¡ë  ìˆ˜ ìˆìœ¼ë‹ˆ ì»¬ëŸ¼ì„ ì œê±°í•©ë‹ˆë‹¤.

```python
X2 = X.drop('cust_id', axis=1)
y2 = y.drop('cust_id', axis=1)
X2, y2
```

- ë°ì´í„°ì— ê²°ì¸¡ì¹˜ê°€ ìˆëŠ”ì§€ í™•ì¸í•´ë³´ì„¸ìš”


```python
X2.isnull().values.any(), y.isnull().values.any()
```

- ê²°ì¸¡ì¹˜ì— 0ìœ¼ë¡œ ì±„ì›Œ ë„£ì–´ ëª¨ë¸ í•™ìŠµì— ì§€ì¥ì´ ì—†ë„ë¡ í•©ë‹ˆë‹¤.

```python
X2 = X2.fillna(0)
X2
```


ë¬¸ìí˜• ë²”ì£¼ ë°ì´í„°ë¥¼ ìˆ«ìë¡œ ë°”ê¾¸ê¸° ìœ„í•œ ì¸ì½”ë”©ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.

```python
for col in X2.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X2[col] = le.fit_transform(X2[col])
```

```python
# object type ì—†ëŠ” ê±° í™•ì¸ ì™„ë£Œ 
X2.dtypes
```

ê° ë°ì´í„°ì— í‘œì¤€í™”ë¥¼ ì ìš©í•˜ì—¬ ë°ì´í„°ì˜ ìŠ¤ì¼€ì¼(í¬ê¸° ì°¨ì´)ì„ ë§ì¶°ì¤ë‹ˆë‹¤.
- í‰ê· ì„ 0, í‘œì¤€í¸ì°¨ë¥¼ 1ë¡œ ë§ì¶°ì„œ â†’ ë°ì´í„°ê°€ ì •ê·œ ë¶„í¬ í˜•íƒœë¡œ ë³€í™˜ë˜ë„ë¡ í•˜ì„¸ìš”

```python
scaler = StandardScaler()
X_scaled =  pd.DataFrame(scaler.fit_transform(X2), columns=X2.columns)
```

# 5-1. ëª¨ë¸ë§ - LogisticRegression

* ë³¸ê²©ì ìœ¼ë¡œ ëª¨ë¸ì„ ì„ ì–¸í•˜ê³  í•™ìŠµì‹œí‚µë‹ˆë‹¤.


í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y2, test_size=0.2, random_state=42)
```

ëª¨ë¸ì„ ì„ ì–¸í•˜ì—¬ ê°ì²´í™”ì‹œí‚µë‹ˆë‹¤.

```python
model = LogisticRegression()
```


ëª¨ë¸ì„ í•™ìŠµ ë°ì´í„°ì— ë§ì¶° í•™ìŠµì‹œí‚µë‹ˆë‹¤.

```python
model.fit(X_train, y_train)
```

# 6-1. ì˜ˆì¸¡ ì„±ëŠ¥ í™•ì¸í•´ë³´ê¸° - LogisticRegression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
y_pred = model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_mat = confusion_matrix(y_test, y_pred)

print("ì •í™•ë„ (Accuracy):", acc)
print("ì •ë°€ë„ (Precision):", precision)
print("ì¬í˜„ìœ¨ (Recall):", recall)
print("F1-score:", f1)
print("Confusion Matrix:\n", conf_mat)
```
