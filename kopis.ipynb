```python
import matplotlib.pyplot as plt

# ▶ 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS
plt.rcParams['axes.unicode_minus'] = False

# ▶ 데이터
regions = ['서울', '경기/인천', '충청도', '강원도', '경상도', '전라도', '제주도']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#6baed6', '#fd8d3c', '#74c476', '#f768a1', '#9e9ac8', '#fdd0a2', '#fdae6b']
total = sum(counts)

# ▶ 내부 텍스트를 위한 커스텀 함수
def make_label(region, count):
    pct = count / total * 100
    return f"{region}\n{count:,}건\n({pct:.1f}%)"

# ▶ 파이 차트
fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(
    counts,
    labels=None,
    colors=colors,
    startangle=140,
    wedgeprops=dict(edgecolor='white'),
    pctdistance=0.6,
    autopct='%1.1f%%'  # 형식은 무시될 예정
)

# ▶ 내부 텍스트를 지역명 + 건수 + 비율로 교체
for i, autotext in enumerate(autotexts):
    label = make_label(regions[i], counts[i])
    autotext.set_text(label)
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')
    autotext.set_color('white')
    autotext.set_ha('center')
    autotext.set_va('center')

# ▶ 제목
ax.set_title('지역별 공연 횟수 비율', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```
