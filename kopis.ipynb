```python
import matplotlib.pyplot as plt

#   
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS
plt.rcParams['axes.unicode_minus'] = False

#  
regions = ['', '/', '', '', '', '', '']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#6baed6', '#fd8d3c', '#74c476', '#f768a1', '#9e9ac8', '#fdd0a2', '#fdae6b']
total = sum(counts)

#      
def make_label(region, count):
    pct = count / total * 100
    return f"{region}\n{count:,}\n({pct:.1f}%)"

#   
fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(
    counts,
    labels=None,
    colors=colors,
    startangle=140,
    wedgeprops=dict(edgecolor='white'),
    pctdistance=0.6,
    autopct='%1.1f%%'  #   
)

#     +  +  
for i, autotext in enumerate(autotexts):
    label = make_label(regions[i], counts[i])
    autotext.set_text(label)
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')
    autotext.set_color('white')
    autotext.set_ha('center')
    autotext.set_va('center')

#  
ax.set_title('   ', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```
