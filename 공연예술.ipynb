```python
import matplotlib.pyplot as plt

# ✅ 한글 폰트 설정 (Windows 사용자: Malgun Gothic, macOS: AppleGothic, Linux: NanumGothic 등)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 또는 'AppleGothic', 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# ✅ 데이터
regions = ['서울', '경기/인천', '충청도', '강원도', '경상도', '전라도', '제주도']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']

# ✅ 막대그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(regions, counts, color=colors)

# ✅ 제목 및 축 라벨
plt.title('지역별 공연건수', fontsize=16)
plt.xlabel('지역', fontsize=12)
plt.ylabel('공연건수', fontsize=12)

# ✅ 각 바 위에 수치 표기
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 100,
             f'{height:,}', ha='center', va='bottom', fontsize=11)

# ✅ 기타 설정
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# ✅ 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
plt.rcParams['axes.unicode_minus'] = False

# ✅ 데이터
regions = ['서울', '경기/인천', '충청도', '강원도', '경상도', '전라도', '제주도']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']

# ✅ 막대그래프 생성
plt.figure(figsize=(13, 13))
bars = plt.bar(regions, counts, color=colors)
plt.ylim(0, max(counts) + 300) 
# ✅ 제목 및 축 라벨 (크고 굵고 검정색)
plt.title('지역별 공연건수', fontsize=24, fontweight='bold', color='black')
plt.xlabel('지역', fontsize=18, fontweight='bold', color='black')
plt.ylabel('공연건수', fontsize=18, fontweight='bold', color='black')

# ✅ 각 바 위에 수치 표기 (크게, 굵게, 검정색)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 100,
             f'{int(height):,}', ha='center', va='bottom',
             fontsize=20, fontweight='bold', color='black')

# ✅ X축 라벨도 굵고 크게
plt.xticks(fontsize=20, fontweight='bold', color='black')
# ✅ Y축 눈금도 굵고 크게
plt.yticks(fontsize=16, fontweight='bold', color='black')

# ✅ 기타 설정
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

regions = ['서울', '경기/인천', '충청도', '강원도', '경상도', '전라도', '제주도']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']

plt.figure(figsize=(12,12))
bars = plt.bar(regions, counts, color=colors)

# ✅ 제목: 크게 + 두껍게 + 여백
plt.title('지역별 공연건수', fontsize=28, fontweight='bold', pad=40, color='black')

# ✅ 축 라벨: 크기 다르게 + 색상 회색으로
plt.xlabel('지역', fontsize=18, fontweight='bold', color='gray', labelpad=15)
plt.ylabel('공연건수', fontsize=18, fontweight='bold', color='gray', labelpad=15)

# ✅ 바 위 숫자
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 150,
             f'{int(height):,}', ha='center', va='bottom',
             fontsize=20, fontweight='bold', color='black')

# ✅ 축 눈금
plt.xticks(fontsize=20, fontweight='bold', color='black')
plt.yticks(fontsize=16, fontweight='bold', color='black')

# ✅ y축 여유 공간
plt.ylim(0, max(counts) + 500)

plt.tight_layout()
plt.show()
```
