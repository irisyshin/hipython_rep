```python
import requests
from bs4 import BeautifulSoup
musical_url = requests.get('https://tickets.interpark.com/contents/ranking?genre=MUSICAL')
musical_soup = BeautifulSoup(musical_url.text, 'html.parser')
```

```python
# title, location, reservation rate 불러오기 
title_list =  musical_soup.select('div.responsive-ranking-list_rankingContentsInner__8FuZE > li')
title_list = [title.text for title in title_list] # save title on list 

location_list = musical_soup.select('div.responsive-ranking-list_rankingContentsInner__8FuZE > div > li')
location_list = [location.text for location in location_list] # save location on list 

rsv_list = musical_soup.select('div.responsive-ranking-list_rankingContentsBottom__7oRHi > li')
rsv_list = [rsv.text for rsv in rsv_list] # save reservation rate on list 

title_list, location_list, rsv_list
```

```python
import pandas as pd
import re

# 1. 제목 정제 함수
def clean_title(title):
    match = re.search(r'〈(.*?)〉', title)
    if match:
        return match.group(1).strip()
    else:
        return title.replace('뮤지컬', '').strip()

# 2. 예약률 정제 함수
def clean_percentage(rate_str):
    return float(rate_str.replace('%', '').strip())

# 3. 정제된 리스트 생성
cleaned_titles = [clean_title(t) for t in title_list]
cleaned_rates = [clean_percentage(r) for r in rsv_list]

# 4. DataFrame 생성
df = pd.DataFrame({
    "Rank": list(range(1, len(cleaned_titles)+1)),
    "Title": cleaned_titles,
    "Location": location_list,
    "ReservationRate": cleaned_rates
})

# 5. CSV 저장
df.to_csv("musicals_cleaned.csv", index=False, encoding="utf-8-sig")
print("CSV 저장 완료: musicals_cleaned.csv")
```
